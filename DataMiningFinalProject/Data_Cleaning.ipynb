### Remove duplicate values

```python
dataset = read_excel(datasetstring)
print(dataset.shape)
dataset.drop_duplicates(inplace=True)
print(dataset.shape)
```

###  Selected to be used Attributes

```python
dataset = dataset[["Book_ISBN", "Book_Name","Book_Publisher","Book_Author", "Book_Price","Book_Page","Book_Publish_Date","Book_Rate", "Book_Rating_Amount","Book_Detail"]]
```

```python
dataset.head()
```

### Converting data to desired shape

```python
#Removing ' pages' part from Book_Page attribute
dataset["Book_Page"] = dataset.Book_Page.str[:-6]
#Removing other stuf to get year of published date of book
dataset["Book_Publish_Date"]=dataset.Book_Publish_Date.str[7:11]
#Removing 'US$' part from Book_Price to make Book_Price float attribute
dataset["Book_Price"]=dataset.Book_Price.str[3:]
#Removing '( ratings by Goodreads)' from Book_Rating_Amount attribute
dataset["Book_Rating_Amount"]=dataset.Book_Rating_Amount.str[1:-21]
#Cleaning Detail attribute  
dataset["Book_Detail"]=dataset.Book_Detail.str[:9]
```

```python
dataset.loc[dataset.Book_Detail == "English", "Book_Detail"] = "Unknown"
dataset.loc[dataset.Book_Detail == "Board boo", "Book_Detail"] = "Board Book"
dataset.loc[dataset.Book_Detail == "Cards\n\n\nE", "Book_Detail"] = "Cards"
dataset.loc[dataset.Book_Detail == "Book\n\nKAK", "Book_Detail"] = "Unknown"
dataset.loc[dataset.Book_Detail == "Hardback\n", "Book_Detail"] = "Hardback"
dataset.loc[dataset.Book_Detail == "Mixed med", "Book_Detail"] = "Mixed"
dataset.loc[dataset.Book_Detail == "Sheet mus", "Book_Detail"] = "Unknown"
dataset.loc[dataset.Book_Detail == "Toy\n\n\nEng", "Book_Detail"] = "Unknown"
dataset.loc[dataset.Book_Detail == "Game\n\n\nEn", "Book_Detail"] = "Unknown"
```

```python
#Comma is converted to dot for not causing any error while converting it to float
dataset['Book_Rating_Amount']=dataset['Book_Rating_Amount'].apply(lambda x:str(x).replace(',','.'))
```

```python
dataset.head()
```

### Converting data types

```python
dataset['Book_Rate'] = dataset['Book_Rate'].astype(float)
```

```python
dataset['Book_Rating_Amount'] = dataset['Book_Rating_Amount'].str.strip().astype(float)
```

```python
dataset['Book_Publish_Date'] = dataset['Book_Publish_Date'].astype(int)
```

```python
price=dataset['Book_Price'] = dataset['Book_Price'].astype(float)
```

```python
dataset = dataset[~dataset['Book_Page'].isnull()]
dataset['Book_Page'] = dataset['Book_Page'].astype(int)
```

```python
print(dataset.info())
```

### Fill the in the blanks

```python
dataset.isnull().sum()
```

### Fill the Unknown of name entity instead of NAN

```python
dataset['Book_Author']=dataset['Book_Author'].fillna('Unknown')
```

### Null values Date fill with first of unique values

```python
dataset['Book_Page']=dataset['Book_Page'].fillna(dataset['Book_Page'].unique()[0])
```

### Empty fields are filled with average and meadian value 

```python
dataset['Book_Price']=dataset['Book_Price'].fillna(dataset['Book_Price'].mean())
dataset['Book_Rating_Amount']=dataset['Book_Rating_Amount'].fillna(dataset['Book_Rating_Amount'].median())
dataset['Book_Rate']=dataset['Book_Rate'].fillna(dataset['Book_Rate'].median())
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
dataset.to_excel(datasetfilledstring, index=False)
```
