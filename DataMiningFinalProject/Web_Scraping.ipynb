# All Books in Best Seller Section

```python
for i in range (1,35):   
    website = 'https://www.bookdepository.com/bestsellers?page='+ str(i)+''
    response=requests.get(website,headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    results =soup.find_all("div",attrs={"class":"book-item"})
    for result in results:
        book_links = result.find_all("div",attrs={"class":"item-info"})
        for book in book_links:
            link = book.find("h3",attrs={"class":"title"})
            link_continue = link.a.get("href")
            link_head = "https://www.bookdepository.com/"
            link_all = link_head + link_continue
            print(link_all)
```

# Fetching HTML data and appends list of book which called 'booklist'

```python
booklist=[]
for i in range (1,34):   
    website = 'https://www.bookdepository.com/bestsellers?page='+ str(i)+''
    response=requests.get(website,headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    results =soup.find_all("div",attrs={"class":"book-item"})
    for result in results:
        book_links = result.find_all("div",attrs={"class":"item-info"})
        for book in book_links:
            link = book.find("h3",attrs={"class":"title"})
            link_continue = link.a.get("href")
            link_head = "https://www.bookdepository.com/"
            link_all = link_head + link_continue

            request = requests.get(link_all, headers = headers )
            response= BeautifulSoup(request.content, "html.parser")

            for detail in response:
                try:
                    book_isbn =  detail.find("span",attrs={"itemprop":"isbn"}).text.strip()
                    print(book_isbn)
                except:
                    book_isbn=''
                try:
                    book_name=detail.find("h1",attrs={"itemprop":"name"}).text.strip()
                    print(book_name)
                except:
                    book_name=''
                try:
                    book_author= detail.find("span",attrs={"itemprop":"author"}).text.strip()
                    print(book_author)
                except:
                    book_author=''
                try:
                    book_publisher = detail.find("span",attrs={"itemprop":"publisher"}).text.strip()
                    print(book_publisher)
                except:
                    book_publisher=''
                try:
                    #Some of books' prices might be discounted so to accurate data I took list-price which means its original price
                    book_price = detail.find("span",attrs={"class":"list-price"}).text.strip()
                    print(book_price)
                except:
                    try:
                         book_price = detail.find("span",attrs={"class":"sale-price"}).text.strip()
                    except:
                        book_price=''
                try:
                    book_page = detail.find("span",attrs={"itemprop":"numberOfPages"}).text.strip()
                    print(book_page)
                except:
                    book_page=''
                try:
                    book_publish_date= detail.find("span",attrs={"itemprop":"datePublished"}).text.strip()
                    print(book_publish_date)
                except:
                    book_publish_date=''
                try:
                    book_detail = detail.find("ul",attrs={"class":"meta-info hidden-md"}).text.strip()
                    print(book_detail)
                except:
                    book_detail=''                
                try:
                    book_rate = detail.find("span",attrs={"itemprop":"ratingValue"}).text.strip()
                    print(book_rate)                    
                except:
                    book_rate=''
                try:
                    book_rating_amount = detail.find("span",attrs={"class":"rating-count"}).text.strip()
                    commacounter= book_rating_amount.count(',')
                    if commacounter > 1 :
                        book_rating_amount =book_rating_amount.replace(',','',1)
                    print(book_rating_amount)
                except:
                    book_rating_amount=''
                booklist.append([book_isbn,book_name,book_publisher,book_author,book_price,book_page,book_publish_date,book_rate,book_rating_amount,book_detail])
```

# Converts booklist to Dataframe

```python
excelfile = pd.DataFrame(booklist,columns = ["Book_ISBN", "Book_Name","Book_Publisher",
                     "Book_Author", "Book_Price","Book_Page",
                     "Book_Publish_Date",
                     "Book_Rate", "Book_Rating_Amount",
                     "Book_Detail"])
#Exports data to excel file.
excelfile.to_excel("Dataset.xlsx",index=False)
```
